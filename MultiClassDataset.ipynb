```python
import pandas as pd
import os
import shutil
```

```python
FILE_PATH ='C:\\Users\\harit\\Downloads\\covid-chestxray-dataset-master\\metadata.csv'
IMAGES_PATH ='C:\\Users\\harit\\Downloads\\covid-chestxray-dataset-master\\images'
```

```python
df = pd.read_csv(FILE_PATH)
df.head()
```

```python
TARGET_DIR_COVID = "MultiClass\\Covid"
TARGET_DIR_VIRAL = "MultiClass\\Viral"
TARGET_DIR_BACTERIAL = "MultiClass\\Bacterial"
TARGET_DIR_FUNGAL = "MultiClass\\Fungal"
TARGET_DIR_NORMAL = "MultiClass\\Normal"
```

```python
TARGET_DIR = "MultiClass"
if not os.path.exists(TARGET_DIR):
    os.makedirs(TARGET_DIR_COVID)
    os.makedirs(TARGET_DIR_VIRAL)
    os.makedirs(TARGET_DIR_BACTERIAL)
    os.makedirs(TARGET_DIR_FUNGAL)
    os.makedirs(TARGET_DIR_NORMAL)
    print("Folder created")
```

```python
cnt=1

for (i,row) in df.iterrows():
    if(row["finding"]=="Pneumonia/Viral/COVID-19" and row["folder"]=='images' and(row["view"]=='PA' or row["view"]=='AP' or row["view"]=='AP Erect' or row["view"]=='AP Supine') and row['modality']=='X-ray'):
        filename = row['filename']
        image_path = os.path.join(IMAGES_PATH,(filename))
        image_copy_path = os.path.join(TARGET_DIR_COVID,filename)
        shutil.copy2(image_path,image_copy_path)
        print("moving image ",cnt)
    elif((row["finding"]=='Pneumonia/Viral/SARS' or row["finding"]=='Pneumonia/Viral/MERS-CoV' or row["finding"] =='Pneumonia/Viral/Varicella' or row["finding"]=='Pneumonia/Viral/Influenza' or row["finding"]=='Pneumonia/Viral/Herpes') and row["folder"]=='images' and(row["view"]=='PA' or row["view"]=='AP' or row["view"]=='AP Erect' or row["view"]=='AP Supine') and row['modality']=='X-ray'):
        filename = row['filename']
        image_path = os.path.join(IMAGES_PATH,(filename))
        image_copy_path = os.path.join(TARGET_DIR_VIRAL,filename)
        shutil.copy2(image_path,image_copy_path)
        print("moving image ",cnt)
    elif((row["finding"]=='Pneumonia/Bacterial/Streptococcus' or row["finding"]=='Pneumonia/Bacterial/Klebsiella' or row["finding"]=='Pneumonia/Bacterial/E.Coli' or row["finding"]=='Pneumonia/Bacterial/Nocardia' or row["finding"]=='Pneumonia/Bacterial/Mycoplasma' or row["finding"]=='Pneumonia/Bacterial/Legionella' or row["finding"]=='Pneumonia/Bacterial/Chlamydophila'or row["finding"]=='Pneumonia/Bacterial/Staphylococcus/MRSA') and row["folder"]=='images' and(row["view"]=='PA' or row["view"]=='AP' or row["view"]=='AP Erect' or row["view"]=='AP Supine') and row['modality']=='X-ray'):
        filename = row['filename']
        image_path = os.path.join(IMAGES_PATH,(filename))
        image_copy_path = os.path.join(TARGET_DIR_BACTERIAL,filename)
        shutil.copy2(image_path,image_copy_path)
        print("moving image ",cnt)
    elif((row["finding"] =='Pneumonia/Fungal/Pneumocystis' or row["finding"]=='Pneumonia/Fungal/Aspergillosis') and row["folder"]=='images' and(row["view"]=='PA' or row["view"]=='AP' or row["view"]=='AP Erect' or row["view"]=='AP Supine') and row['modality']=='X-ray'):
        filename = row['filename']
        image_path = os.path.join(IMAGES_PATH,(filename))
        image_copy_path = os.path.join(TARGET_DIR_FUNGAL,filename)
        shutil.copy2(image_path,image_copy_path)
        print("moving image ",cnt)
    cnt+=1
```
